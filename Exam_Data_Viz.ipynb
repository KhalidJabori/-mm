### <span style="color:blue"> <span  style="font-family:Times New Roman"> Divisons maintenant les données globales pour tirer des conclusions plus précises. Créez un nouveau Dataframe, où le profit est négatif, ce qui signifie une perte, et concentrez-vous sur ces domaines à améliorer. </span> </span>

```python
#Cela créera un nouveau df avec toutes les fonctionnalités où le profit est inférieur à zéro.
loss_df=data[data['Profit'] < 0]
loss_df
```

```python
#Dimension du Dataframe loss_df
loss_df.shape
```

### <span style="color:#00A86B"> <span  style="font-family:Cambria"> Cette perte df contient 1871 lignes et 13 colonnes. </span> </span>

```python
# Afficher la description statistique de la Dataframe de perte.
loss_df.describe().T
```

```python
Total_loss=np.negative(loss_df['Profit'].sum())
print("Perte totale  = %.2f" %Total_loss)
```

```python
loss_df.groupby(by='Segment').sum()
```
